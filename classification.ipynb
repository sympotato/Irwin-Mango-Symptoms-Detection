```python
import cv2
import pandas as pd
import os
import tensorflow as tf
```

```python
#path= os.path.abspath(os.path.dirname(__file__))
train = pd.read_csv("/home/user/gift/train1.csv",encoding='utf8').fillna('null')
pic_path='/home/user/Downloads/C2_TrainDev/Train/'
a=[]
for i in range(len(train)):
    a.append(train.loc[i][0])
    b=[]
    for j in range(1,28):
        if train.loc[i][j*5] not in b:
            if train.loc[i][j*5]=='null':
                continue
            b.append(int(train.loc[i][j*5]))
    b.sort()
    c=''
    for k in b:
        c=c+str(k)
    d=[]
    d.append(c)
    #path = os.path.abspath(os.path.dirname(__file__))
    path_data = "/home/user/gift" + '/pic512/' + d[0]
    if not os.path.exists(path_data):
        os.makedirs(path_data)
    pic = cv2.imread(pic_path+train.loc[i][0])
    pic = cv2.resize(pic, (512, 512), interpolation=cv2.INTER_AREA)
    cv2.imwrite(path_data + "/" + train.loc[i][0],pic) 
```
