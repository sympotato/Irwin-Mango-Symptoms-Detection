```python
import os, random, shutil

def copyFile(fileDir):
    pathDir = os.listdir(fileDir)
    sample = random.sample(pathDir, 1000)
    for name in sample:
        shutil.copyfile(fileDir+"/"+name, tarDir+"/"+name)

def copyFile2(fileDir):
    pathDir = os.listdir(fileDir)
    sample = random.sample(pathDir, 250)
    for name in sample:
        shutil.copyfile(fileDir+"/"+name, tarDir+"/"+name)
```
