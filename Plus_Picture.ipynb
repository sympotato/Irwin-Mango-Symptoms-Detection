```python
# -*- coding:utf-8 -*-
import os, random, shutil
import os.path
import tensorflow
import cv2
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
path = '/home/user/gift/more_1000/'    #獲取當前路徑

datagen = ImageDataGenerator(
                    zca_whitening=False,
                    rotation_range=40,
                    width_shift_range=0.2,
                    height_shift_range=0.2,
                    shear_range=0.2,
                    zoom_range=0.2,
                    horizontal_flip=True,
                    fill_mode='nearest')

num_files = 0 #路徑下檔案數量(包括資料夾)
files_name=[]
```

```python
for fn in os.listdir(path):
        num_files += 1
        #print(fn)
        files_name=files_name+[fn]

print(num_files)
print(files_name)
print(len(files_name))
```

```python
for i in range(0,len(files_name)):
        pic_path = path  + files_name[i]+"/"
        num_pic = 0
        pic_name=[]
        for fn in os.listdir(pic_path):
                num_pic += 1
                pic_name=pic_name+[fn]
        T=(1500//num_pic)   #數字部分是需要的張數*1.5
        print(T)
        print(num_pic)
        print(pic_path)
        for k in range(num_pic):
            L=0;
            img = cv2.imread(pic_path+pic_name[k])
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  #格式必須為：(sample數, channels, height, width)
            img = img.reshape((1,) + img.shape)
            if(T>=1):
                for batch in datagen.flow(img, batch_size=1,save_to_dir=pic_path, save_prefix='', save_format='jpg'):
                    L += 1
                    if L >= T:
                        break
```
